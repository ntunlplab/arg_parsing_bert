```python
import json
```

```python
data = []
with open('./../preprocess/total_eval') as f:
    for line in f:
        data.append(json.loads(line))
```

```python
datas = []
for d in data:
    for context, bio in zip(d['context'], d['bio']):
        datas.append([context,d['topic'],bio])
```

```python
len(datas)
```

```python
ans = [[],[]]
with open('./saved_models/total_crf/pred.txt') as f:
    for i, line in enumerate(f):
        ans[i%2].append(line.strip())
```

```python
temp = []
count = 0
ana = {}
for pred, label, context in zip(ans[0], ans[1], datas):
    if(pred!=label):
        total = []
        t_a, t_b = 0, 0
        for p, l, c in zip(pred.split(), label.split(), context[0].split()):
            if(p!=l):
                total.append(c)
            if(p=='B'):
                t_a+=1
            if(l=='B'):
                t_b+=1
        try:
            ana[t_b-t_a]+=1
        except:
            ana[t_b-t_a]=1
        if(t_a==t_b):
            continue
        if(len(total)==1 and total[0]=='.'):
            continue
        temp.append([pred,label,context])
        if(t_b-t_a == 1):
            print(pred)
            print(label)
            print(context[0])
            print()
            print(context[1])
            print(total)
            print('-------------------------------------')
```
